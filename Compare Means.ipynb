```python
### Chi-squared test comparing the Sex of COVID-19 positive individuals with those who are negative

import pandas as pd
from scipy import stats
from scipy.stats import chi2_contingency
import numpy as np

data = pd.read_csv(r'D:\DATA\Python\2\Data.csv')

# Create a contingency table
contingency_table = pd.crosstab(data['Sex'], data['Target'])

# Perform Chi-squared test
chi2, pval, dof, expected = chi2_contingency(contingency_table)

# Print the results
print("Chi-squared test results for Sex:")
print("Chi-squared statistic:", format(chi2, '.2f'))
# print("p-value:", pval)
print("p-value:", format(pval, '.2f'),"\n ----------------------------")
print("\n Observed frequencies:\n", contingency_table)


import numpy as np
from np import array

# define your table as a 2D numpy array
# provide a way to efficiently store and manipulate large amounts of data.
table = array(contingency_table)

# find the sum of every column
# axis=0 for sum of every coulmn and axis=1 for sum of every row
column_sums = np.sum(table, axis=0)
Percentage_frequency=(contingency_table /column_sums)*100


# print("\nPercentage_frequency:\n",Percentage_frequency)
# Round the values in the array to 2 decimal places.
rounded_Percentage = np.array(Percentage_frequency)
for row in rounded_Percentage:
    for i in range(len(row)):
        row[i] = round(row[i], 2)
print("\nPercentage_frequency:\n",rounded_Percentage,"\n ----------------------------")

# Total
row_sums = np.sum(table, axis=1)
print("\nTotal:\n",row_sums)

# Round the values in the array to 2 decimal places.
Percentage_frequencyN=(row_sums /sum(row_sums))*100
PercentageN = np.array(Percentage_frequencyN)
rounded_Percentage = np.round(PercentageN, decimals=2)
print(rounded_Percentage)
```

```python
### Leven test and Mannwhitney test comparing the Blood of COVID-19 positive individuals with those who are negative


from scipy.stats import levene
from scipy.stats import mannwhitneyu
from scipy.stats import sem
from numpy import mean

# Create a contingency table
data = pd.read_csv(r'D:\DATA\Python\2\Data.csv')
# Introduce groups
Group1 = data.loc[data['Target'] == 'No'] 
Group1List = list(Group1['Blood'])
Group2 = data.loc[data['Target'] == 'Yes'] 
Group2List = list(Group2['Blood'])

# Levene test
stat0, p_value0 = levene(Group1List, Group2List)
print('p-value levene test:', p_value0,"\n ----------------------------")

# Mannwhitney test
stat1, p_value1 = mannwhitneyu(Group1List, Group2List)
print('stat:', round(stat1,2))
print('p-value:', round(p_value1,2))

mean1 = mean(Group1List)
mean2 = mean(Group2List)
se1 = sem(Group1List)
se2 = sem(Group2List)

print("mean1:",round(mean1,2),"\n","mean2:",round(mean2,2))
print("Standard error1:",round(se1,2),"\n","Standard error2:",round(se2,2),"\n ----------------------------")

# Total
# extract the column of interest into a numpy array
data = data['Blood'].to_numpy()
mean = mean(data)
se = sem(data)
print("mean:",round(mean,2),"\n", "Standard error:", round(se,2))




```

```python
### Leven test and T test comparing the biomarker_1 of COVID-19 positive individuals with those who are negative
import pandas as pd
from scipy import stats
from scipy.stats import levene
import numpy as np

data = pd.read_csv(r'D:\DATA\Python\2\Data.csv')
# Introduce groups
Group1 = data.loc[data['Target'] == 'No'] 
Group1List = list(Group1['biomarker_1'])
Group2 = data.loc[data['Target'] == 'Yes'] 
Group2List = list(Group2['biomarker_1'])

# Levene test
stat0, p_value0 = levene(Group1List, Group2List)
print('p-value levene test:', p_value0,"\n ----------------------------")

# T test
t_statistic, p_value = stats.ttest_ind(Group1List, Group2List)

print('t-statistic:', round(t_statistic,2))
print('p-value:', round(p_value,2))


mean1 = np.mean(Group1List)
mean2 = np.mean(Group2List)
se1 = stats.sem(Group1List)
se2 = stats.sem(Group2List)

print("mean1:",round(mean1,2),"\n","mean2:",round(mean2,2))
print("Standard error1:",round(se1,2),"\n","Standard error2:",round(se2,2),"\n ----------------------------")

# Total
# extract the column of interest into a numpy array
data = data['biomarker_1'].to_numpy()
mean = np.mean(data)
se = stats.sem(data)
print("mean:",round(mean,2),"\n", "Standard error:", round(se,2))



# create 95% confidence interval
data = np.array(data)
ci=stats.t.interval(confidence=0.95, df=len(data)-1,loc=mean,scale=se)
print ("ci=",ci)



```
